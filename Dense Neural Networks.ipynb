Model Architecture:

2. Dense --> number of nodes: 500, ActivationFunction: relu
3. Dense --> number of nodes: 300, ActivationFunction: relu
4. Dropout ---> argument: 0.3

```python
#1. Import Libraries
```

```python
import cv2 as cv
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt 
import tensorflow.keras.layers as layers
import tensorflow.keras.models as models
import tensorflow.keras.datasets.mnist as mnist
```

```python
#2. Load the Data
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
X_train.shape
```

```python
#3. Normalize the images
```

```python
X_train, X_test = X_train/255.0 , X_test/255.0
```

```python
#4. Model Architecture (according to the mentioned architecture above!)
```

```python
model = models.Sequential([
    layers.Flatten(input_shape = (28, 28)),
    layers.Dense(500, activation='relu'),
    layers.Dense(300, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(10, activation='softmax')    
    
])
```

```python
#5. Compile
```

```python
model.compile( loss = 'sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
#6. Fit
```

```python
hist = model.fit(X_train, y_train,validation_split=0.3, epochs=10,batch_size=100, verbose=1)
```

```python
#7. Evaluate
```

```python
model.evaluate(X_test, y_test, verbose=1)
```

```python
hist.history.keys()
```

```python
plt.plot(hist.history['accuracy'])
plt.plot(hist.history['val_accuracy'])
```
